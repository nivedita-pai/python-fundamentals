# Lists in Python
A list is an ordered, mutable collection of items.
- Ordered — items have a fixed position
- Mutable — items can be changed after creation
- Allows duplicate values
- Can hold different datatypes

```python
# Empty list
my_list = []

# List of integers
numbers = [1, 2, 3, 4, 5]

# List of strings
fruits = ["apple", "banana", "mango"]

# Mixed datatypes
mixed = [1, "hello", 3.14, True]

print(numbers)
print(fruits)
print(mixed)
print(type(mixed))
```

## Indexing and Slicing
Same as strings — index starts from 0!

```python
fruits = ["apple", "banana", "mango", "grape", "orange"]

print(fruits[0])    # first item
print(fruits[-1])   # last item
print(fruits[1:3])  # slicing
print(fruits[::-1]) # reversed
```

## Adding Items

```python
fruits = ["apple", "banana", "mango"]

# append - adds to the end
fruits.append("grape")
print(fruits)

# insert - adds at specific index
fruits.insert(1, "orange")
print(fruits)
```

## Removing Items

```python
fruits = ["apple", "banana", "mango", "grape"]

# remove - removes by value
fruits.remove("banana")
print(fruits)

# pop - removes by index (default last item)
fruits.pop()
print(fruits)

# pop at specific index
fruits.pop(0)
print(fruits)
```

## Sorting and Reversing

```python
numbers = [5, 2, 8, 1, 9, 3]
numbers.reverse() #reverse the list
print(numbers)

numbers.sort() #ascending order
print(numbers)

numbers.sort(reverse=True) #descending order
print(numbers)

numbers.reverse() #reverse the list again to get original order
print(numbers)
```

## Other Useful Methods

```python
numbers = [1, 2, 3, 2, 4, 2, 5]

# count - counts occurrences of a value
print(numbers.count(2))

# index - returns index of first occurrence
print(numbers.index(2))

# len - total items in list
print(len(numbers))

# in - check if item exists
print(3 in numbers)
print(10 in numbers)
```

## Lists are Mutable
Unlike strings, lists can be modified in place!

```python
fruits = ["apple", "banana", "mango"]

# Directly change an item
fruits[1] = "orange"
print(fruits)
```
