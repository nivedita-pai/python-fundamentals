# Tuples in Python
A tuple is an ordered, immutable collection of items.
- Ordered — items have a fixed position
- Immutable — cannot be changed after creation
- Allows duplicate values
- Faster than lists

```python
# Empty tuple
my_tuple = ()

# Tuple of integers
numbers = (1, 2, 3, 4, 5)

# Tuple of strings
fruits = ("apple", "banana", "mango")

# Mixed datatypes
mixed = (1, "hello", 3.14, True)

print(numbers)
print(fruits)
print(mixed)
print(type(numbers))
```

## Indexing and Slicing
Same as lists!

```python
fruits = ("apple", "banana", "mango", "grape", "orange")

print(fruits[0])
print(fruits[-1])
print(fruits[1:3])
print(fruits[::-1])
```

## Tuples are Immutable
Once created, items cannot be changed!

```python
fruits = ("apple", "banana", "mango")

# This will throw a TypeError!
fruits[1] = "orange"
```

## Tuple Methods
Tuples have only 2 methods since they are immutable!

```python
numbers = (1, 2, 3, 2, 4, 2, 5)

# count - counts occurrences of a value
print(numbers.count(2))

# index - returns index of first occurrence
print(numbers.index(2))
```

## When to use Tuple over List?
- When data should not change (days of week, months, coordinates)
- Tuples are faster than lists
- Used as dictionary keys (lists cannot be used as keys!)

```python
# Good use cases for tuples
days = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
coordinates = (13.0827, 80.2707)  # Chennai coordinates!
rgb = (255, 0, 0)  # Red color

print(days)
print(coordinates)
print(rgb)
```
