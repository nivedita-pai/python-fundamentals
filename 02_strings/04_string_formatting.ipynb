# String Formatting
Different ways to format and combine strings in Python.

## Concatenation
Joining strings using + operator.

```python
name = "Nivedita"
age = 25

# This works
print("My name is " + name)

# This will throw an error!
print("My age is " + age)
```

```python
name = "Nivedita"
age = 40

# Fix - convert int to str first
print("My name is " + name + " and I am " + str(age) + " years old")
```

## format() method
A cleaner way to format strings using placeholders {}.

```python
name = "Nivedita"
age = 40

print("My name is {} and I am {} years old".format(name, age))

# Can also use index
print("My name is {0} and I am {1} years old".format(name, age))

# Can also use variable names
print("My name is {n} and I am {a} years old".format(n=name, a=age))
```

## f-strings (Most popular!)
The cleanest and most modern way to format strings.
Introduced in Python 3.6

```python
name = "Nivedita"
age = 40

print(f"My name is {name} and I am {age} years old")

# Can even do expressions inside!
print(f"Next year I will be {age + 1} years old")
print(f"Name in uppercase: {name.upper()}")
```
