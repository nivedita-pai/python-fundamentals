# String Indexing and Slicing
Every character in a string has a position called an index.
Indexing starts from 0 in Python.

```python
name = "Nivedita"
#       N  i  v  e  d  i  t  a
#       0  1  2  3  4  5  6  7

print(name[0])   # N
print(name[1])   # i
print(name[7])   # a
```

## Negative Indexing
You can also access characters from the end using negative indexes.

```python
name = "Nivedita"
#       N  i  v  e  d  i  t  a
#      -8 -7 -6 -5 -4 -3 -2 -1

print(name[-1])   # a
print(name[-3])   # i
print(name[-8])   # N
```

## Slicing
Extract a portion of a string using start and end index.
Syntax: string[start:end]
Note: start is inclusive, end is exclusive!

```python
name = "Nivedita"

print(name[0:4])   # Nive
print(name[4:8])   # dita
print(name[2:6])   # vedi

# Leaving start or end empty
print(name[:4])    # Nive - from beginning
print(name[4:])    # dita - till end
print(name[:])     # Nivedita - entire string
```

## Step Slicing
Add a third parameter to skip characters.
Syntax: string[start:end:step]

```python
name = "Nivedita"

print(name[::2])    # every 2nd character
print(name[::1])    # every character (default)
print(name[::-1])   # reverse the string!
```

## Index out of range
Accessing an index that doesn't exist throws an IndexError!

```python
name = "Nivedita"

# This will throw an error!
print(name[10])
```

## IndexError
Python throws an IndexError when you try to access an index that doesn't exist.
"Nivedita" has only 8 characters — valid indexes are 0 to 7!
Always make sure your index is within range.
