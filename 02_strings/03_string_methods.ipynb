# String Methods
Python has many built-in methods to work with strings.
Syntax: string.method_name()

## Case Methods

```python
name = "nivedita pai"

print(name.upper())       # all uppercase
print(name.lower())       # all lowercase
print(name.title())       # first letter of each word capitalized
print(name.capitalize())  # only first letter capitalized
```

## Strip Methods
Used to remove unwanted spaces or characters.

```python
name = "   Nivedita   "

print(name.strip())    # removes spaces from both sides
print(name.lstrip())   # removes spaces from left
print(name.rstrip())   # removes spaces from right
```

## replace() and find()

```python
sentence = "I love Java"

# replace
print(sentence.replace("Java", "Python"))

# find - returns index of first occurrence
print(sentence.find("Java"))

# find returns -1 if not found
print(sentence.find("C++"))
```

## split() and join()

```python
# split - breaks string into a list
sentence = "I am learning Python"
words = sentence.split()
print(words)

# split by specific character
date = "2024-01-15"
parts = date.split("-")
print(parts)

# join - joins a list into a string
words = ["I", "love", "Python"]
sentence = "-".join(words)
print(sentence)
```

## count() , startswith() and endswith()

```python
sentence = "I love Python and Python loves me"

# count - counts how many times a substring appears
print(sentence.count("Python"))

# startswith
print(sentence.startswith("I"))
print(sentence.startswith("Python"))

# endswith
print(sentence.endswith("me"))
print(sentence.endswith("Python"))
```
