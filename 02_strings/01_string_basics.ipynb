# Strings in Python
A string is a sequence of characters enclosed in quotes.

```python
# Single quotes
a = 'Hello'

# Double quotes
b = "World"

# Both are same!
print(a)
print(b)
print(type(a))
```

## Triple Quotes
Used for multiline strings.

```python
message = """This is
a multiline
string!"""

print(message)
```

## len() function
Returns the total number of characters in a string including spaces.

```python
name = "Nivedita"
print(len(name))

sentence = "I am learning Python"
print(len(sentence))
```

## Strings are immutable
Once created, a string cannot be changed.

```python
name = "Nivedita"

# This will throw a TypeError!
name[0] = "K"
```

## Why immutable?
Strings cannot be modified in place.
If you want to change a string, you create a new one!

```python
name = "Nivedita"

# Cannot modify, but can create a new string
new_name = "K" + name[1:]
print(new_name)
```
