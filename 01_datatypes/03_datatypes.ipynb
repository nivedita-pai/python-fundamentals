# Python Datatypes
A datatype defines what kind of value a variable can hold.

Python has the following built-in datatypes:
- int — whole numbers
- float — decimal numbers
- str — text
- bool — True or False
- list, tuple, set, dict — collections (covered separately)

## int
Whole numbers, positive or negative, without decimals.

```python
a = 10
b = -5
c = 0

print(a, b, c)
print(type(a))
```

## float
Numbers with decimal points.

```python
a = 3.14
b = -2.5
c = 0.0

print(a, b, c)
print(type(a))
print(type(b))
print(type(c))
```

## str
Text enclosed in single or double quotes.

```python
name = "Nivedita"
city = 'Chennai'
sentence = "I am learning Python!"

print(name)
print(city)
print(sentence)
print(type(name))
```

## bool
Represents True or False values.

```python
is_student = True
is_working = False

print(is_student)
print(is_working)
print(type(is_student))
```

## Type Conversion
Converting one datatype to another.

```python
# int to float
a = 10
print(float(a))

# float to int
b = 3.14
print(int(b))

# int to string
c = 100
print(str(c))
print(type(str(c)))

# string to int
d = "50"
print(int(d))
```

```python
## What happens when conversion fails?
```

```python
# This will throw an error!
x = "hello"
print(int(x))
```

Python throws a ValueError when conversion is not possible.
"hello" cannot be converted to int because it is not a number!
