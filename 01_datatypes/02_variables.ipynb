# Variables in Python
A variable is a container that stores a value.
In Python, you don't need to declare the type — just assign a value!

```python
name = "Nivedita"
age = 25
height = 5.4
is_student = True

print(name)
print(age)
print(height)
print(is_student)
```

## Variable Naming Rules
- Must start with a letter or underscore
- Cannot start with a number
- Cannot use special characters like @, #, $
- Case sensitive (name and Name are different!)

```python
# Valid variable names
my_name = "Nivedita"
_age = 25
firstName = "Niv"

# Case sensitive
name = "Nivedita"
Name = "Python"
print(name)
print(Name)
```

## Multiple Assignment
Python allows assigning values to multiple variables in one line!

```python
# Assign same value to multiple variables
x = y = z = 10
print(x, y, z)

# Assign different values in one line
a, b, c = 1, 2, 3
print(a, b, c)

# Swap two variables
a, b = b, a
print(a, b)
```

## type() function
Use type() to check the datatype of a variable.

```python
name = "Nivedita"
age = 25
height = 5.4
is_student = True

print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```
