# Hello World in Python
Welcome to Python! Let's write our very first program.

In Java your first program looked like this:
public class Main {
    public static void main(String[] args) {
        System.out.println("Hello World");
    }
}

In Python it's just one line! 🎉

```python
print("Hello World!")
```

## Comments in Python
Comments are lines that Python ignores. Used to explain your code.

```python
# This is a single line comment
print("Comments make code readable!")

'''
This is a multi-line comment
It can span multiple lines
'''

# In Java we use // for single line comments
# In Python we use # for single line comments
```

## print() function
print() is used to display output. Unlike Java, no semicolon needed!

```python
print("Hello World")
print("My name is Nivedita")
print("I am learning Python!")

# printing multiple values
print("Hello", "World", "Python")
```

## input() function
input() is used to take input from the user.

```python
name = input("Enter your name: ")
print("Hello", name)
```
